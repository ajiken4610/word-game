```python
#@title << 2. ゲームを実行
import json
import random
import math

class ConstantValues:
  DIFFICULTY_COUNT = [7,10,15]
  NORMAL_PRICE = 10
  HARD_PRICE = 25
  DIFFICULTY_NAME = ["EASY","NORMAL","HARD"]
  QUESTION_CHOICE_COUNT = 4

class Inputer:
  def inputInteger(message = "整数 ?>",min = 1, max = 4):
    while(True):
      result = input(message)
      try:
        ret = int(result)
        if(ret<min or ret>max): raise ValueError("範囲外")
        return int(result)
      except ValueError:
        print("%dから%dまでの整数で入力!" % (min,max))


class Selector:
  def select(title,message = "Select ?>", count=3):
    print(title)
    ret = Inputer.inputInteger(message,1,count)
    return ret

class Question:
  def __init__(this,src):
    length = len(src)
    keys = list(src)
    this.question = keys[math.floor(random.random()*length)]
    this.correct = math.floor(random.random()*ConstantValues.QUESTION_CHOICE_COUNT)
    this.choice = [0] * ConstantValues.QUESTION_CHOICE_COUNT
    for i in range(ConstantValues.QUESTION_CHOICE_COUNT):
      if(this.correct == i):
        this.choice[i] = src[this.question]
      else:
        this.choice[i] = src[keys[math.floor(random.random()*length)]]
  def ask(this,index,saveData):
    ret = "\n第%d問: %s\n" % (index,this.question)
    for i in range(ConstantValues.QUESTION_CHOICE_COUNT):
      ret += "\t%d. %s\n" % (i+1,this.choice[i])
    userAns = Selector.select(ret,"Answer ?>",4)
    result = userAns == this.correct+1
    print("正解！" if result else "不正解！")
    if(not result):
      if this.choice[this.correct] in saveData.getWrongs():
        saveData.getWrongs()[this.question] +=1
      else:
        saveData.getWrongs()[this.question] = 1
    return result

class Score:
  def __init__(this,initValue = 0):
    this.value = initValue;
  def inc(this):
    this.value += 1
  def dec(this):
    this.value-=1
  def printWith(this,template="Score: %d"):
    print(template % (this.value))

class SaveData:
  def parseFrom(text):
    return SaveData(json.loads(text));
  def __init__(this,saved = {"score":0,"wrongs":{},"having":{}}):
    this.saved = {
        "score":Score(saved["score"]),
        "wrongs":saved["wrongs"],
        "having":saved["having"]
        }
  def getScore(this):
    return this.saved["score"]
  def getWrongs(this):
    return this.saved["wrongs"]
  def getHaving(this):
    return this.saved["having"]
  def toString(this):
    return json.dumps({"score":this.saved["score"].value,"wrongs":this.saved["wrongs"],"having":this.saved["having"]})


class QuestionChain:
  def __init__(this,length):
    this.length = length
  def ask(this,src,saveData):
    score = saveData.getScore()
    for i in range(this.length):
      score.printWith("現在のスコア: %d")
      if Question(src).ask(i + 1,saveData): score.inc()
    score.printWith("現在のスコア: %d")
    return score

class HavingItems:
  def __init__(this,saveData):
    this.saveData = saveData
  def add(this,what):
    this.saveData.getHaving()[what] = True
  def hasItem(this,what):
    return (what in this.saveData.getHaving()) and\
            this.saveData.getHaving()[what]

class Title:
  def start(this):
    return Selector.select("[TITLE]\nコマンドを選択...\n\t1. %s\n\t2. %s\n\t3. %s\n\t4. SHOP\n\t5. EXIT" % (ConstantValues.DIFFICULTY_NAME[0],ConstantValues.DIFFICULTY_NAME[1],ConstantValues.DIFFICULTY_NAME[2]),"1..5 ?>",5)

class EasyQuestionChain(QuestionChain):
  def __init__(this):
    super().__init__(ConstantValues.DIFFICULTY_COUNT[0])
class NormalQuestionChain(QuestionChain):
  def __init__(this):
    super().__init__(ConstantValues.DIFFICULTY_COUNT[1])
class HardQuestionChain(QuestionChain):
  def __init__(this):
    super().__init__(ConstantValues.DIFFICULTY_COUNT[2])
class Shop:
  def __init__(this,save):
    this.save = save
  def ask(this):
    having = HavingItems(this.save)
    while(True):
      result = Selector.select("[SHOP]\t\n1. NORMALを開放\t\n2. HARDを開放\t\n3. EXIT","1..3 ?>",3)
      if result == 1:
        # normal
        if having.hasItem("normal"):
          print("すでに開放済みです。")
        elif this.save.getScore().value < ConstantValues.NORMAL_PRICE:
          print("経験が足らないぞ！")
        else:
          this.save.getScore().value -= ConstantValues.NORMAL_PRICE
          having.add("normal")
          print("NORMALを開放しました。")
      elif result == 2:
        # normal
        if having.hasItem("hard"):
          print("すでに開放済みです。")
        elif not having.hasItem("normal"):
          print("難易度NORMALを解放してからまたおいで")
        elif this.save.getScore().value < ConstantValues.HARD_PRICE:
          print("経験が足らないぞ！")
        else:
          this.save.getScore().value -= ConstantValues.HARD_PRICE
          having.add("hard")
          print("HARDを開放しました。")
      elif result == 3:
        return

class Game:
  def __init__(this,lastSave):
    this.save = SaveData.parseFrom(lastSave) if len(lastSave)!=0 else SaveData()
  words = {}
  def setWords(this,words):
    this.words = words
  def start(this):
    while(True):
      titleResult = Title().start()
      if(titleResult == 1):
        EasyQuestionChain().ask(this.words,this.save)
      elif(titleResult == 2):
        if(HavingItems(this.save).hasItem("normal")):
          NormalQuestionChain().ask(this.words,this.save)
        else:
          print("難易度Normalは開放されていません。SHOPで開放！")
      elif(titleResult == 3):
        if(HavingItems(this.save).hasItem("hard")):
          HardQuestionChain().ask(this.words,this.save)
        else:
          print("難易度Hardは開放されていません。SHOPで開放！")
      elif(titleResult == 4):
        Shop(this.save).ask()
      elif(titleResult == 5):
        break
  def toString(this):
    return this.save.toString()

game = Game(input("セーブデータを入力。ないならそのままEnter?>"));
game.setWords(words)
game.start()
game.toString()
```
